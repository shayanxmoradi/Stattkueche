```python
%load_ext cuml.accel
%load_ext cudf.pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('clear.csv')
df =pd.DataFrame(data)
aggregation_rules = {
    # ویژگی‌های عددی اصلی
    'OrderQty': 'sum',
    'CanceledQty': 'sum',
}

# 4. اجرای تجمیع کامل
df = df.groupby('DateOfOrder').agg(aggregation_rules)
df
```

```python
import pandas as pd
import plotly.express as px


# 3) رسم خطی روی رشتهٔ تاریخ
fig = px.line(df,
              y=['OrderQty'],
              title='تعداد سفارش در طول زمان')

# 4) فرمت محور X
fig.update_xaxes(tickformat='%Y-%m-%d',
                tickangle=-45)

fig.show()
```

```python
# 3) رسم خطی روی رشتهٔ تاریخ
fig = px.line(df,
              y=['CanceledQty'],
              color_discrete_sequence=['#ff4c4c'],
              title='تعداد کنسلی در طول زمان')

# 4) فرمت محور X
fig.update_xaxes(tickformat='%Y-%m-%d',
                tickangle=-45)

fig.show()
```
