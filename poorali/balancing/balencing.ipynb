```python


import pandas as pd
import numpy as np

df = pd.read_csv('/Users/shayan/Desktop/IDS2/Stattkueche/df_weather2.csv')

df_gt5 = df[df['cancel_timing'] == '>5d']
df_2_5 = df[df['cancel_timing'] == '2-5d']
df_le1 = df[df['cancel_timing'] == '<=1d']

n_needed = len(df_gt5) + len(df_2_5) + len(df_le1)
df_no_cancel = (
    df[df['cancel_timing'] == 'no_cancel']
      .sample(n=n_needed, random_state=42)
)

df_4class = pd.concat([df_no_cancel, df_gt5, df_2_5, df_le1], ignore_index=True)

out_path = '/Users/shayan/Desktop/IDS2/Stattkueche/poorali/balancing/4class.csv'
df_4class.to_csv(out_path, index=False)
print(f"✅ Built 4 class dataset with {len(df_4class)} rows; downloading…")



```
